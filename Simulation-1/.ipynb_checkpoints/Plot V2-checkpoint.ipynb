```python
import torch
N = 40
p = 4
r1 = 4
r2 = 3
r3 = 2

resultReal = torch.load('resultReal.py',map_location='cpu')

U2 = torch.tensor(resultReal["Param_Dict"][0]["U2T"].permute(1,0)).reshape(1,N,r2)
for i in range(len(resultReal["Param_Dict"])-1):
    U2 = torch.cat([U2, resultReal["Param_Dict"][i+1]["U2T"].permute(1,0).reshape(1,N,r2)], dim=0)
    
U3 = torch.tensor(resultReal["Param_Dict"][0]["U3T"].permute(1,0)).reshape(1,p,r3)
for i in range(len(resultReal["Param_Dict"])-1):
    U3 = torch.cat([U3, resultReal["Param_Dict"][i+1]["U3T"].permute(1,0).reshape(1,p,r3)], dim=0)
    
U1 = torch.tensor(resultReal["Param_Dict"][0]["U1"]).reshape(1,N,r1)
for i in range(len(resultReal["Param_Dict"])-1):
    U1 = torch.cat([U1, resultReal["Param_Dict"][i+1]["U1"].reshape(1,N,r1)], dim=0)

G1 = torch.tensor(resultReal["Param_Dict"][0]["G1"]).reshape(1,p,r2*r3)
for i in range(len(resultReal["Param_Dict"])-1):
    G1 = torch.cat([G1, resultReal["Param_Dict"][i+1]["G1"].reshape(1,p,r2*r3)], dim=0)

```

```python
def standz(x):
    return((x-torch.mean(x))/torch.std(x))
```
